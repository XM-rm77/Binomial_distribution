```python
from random import random,randint
from scipy.stats import binom
import matplotlib.pyplot as plt
print('Put the number of trials')
m = int(input()) # number of trials
print('Put the number of draws in each trial')
n = int(input()) 
dist_list = [] #Empty list for distribution
for j in range(m):
  success = 0 # setting a zero for counter success in before each trial
  for i in range(n):
    x = randint(0,10) # getting ball from 10 balls 
    if(x >= 3): 
      success += 1 # counting successes
  dist_list.append(success) # Adding number of successes to list

plt.hist(dist_list, bins = 20)
```
